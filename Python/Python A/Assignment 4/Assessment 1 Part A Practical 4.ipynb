# Question 1
1)  A function that takes a NumPy array of integer numbers and returns a NumPy array of
the cube of each the array elements. Your function must use the NumPy universal
function power(). Note you must NOT use for loops to access NumPy array elementsâ€‹

```python
import numpy as np

def numpyCube(nparray):
    return np.power(nparray,3) #return the cube of the nparray using power() function

#check the result
numpy_cube_result = numpyCube(np.arange(10))
print (numpy_cube_result)
    
```

2) A function that takes a Python sequence (range) of integer numbers and returns a python list of the cube of each of the range elements using the Python pow() function.

```python
def rangeCube(rangearray):
    result = [] #create a empty list
    for x in rangearray: #loop through the element
        result.append(pow(x,3)) #append the cube of each elemen into the result list
    return result

#check the result
range_cube_result = rangeCube(range(10))
print(range_cube_result)
```
