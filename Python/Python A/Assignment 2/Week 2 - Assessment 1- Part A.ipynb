# Blood Glucose Data
This section is to create the blood glucose level data of 500 persons without diabetes and the normal values are ranged from 60 mg/dL to 126 mg/dL.

```python
import numpy as np

#create a person_BGL array with 500 data in floating value lie bewteen 60mg/dL - 128 mg/dL
person_BGL_mgdL = np.random.randint(60,128,500).astype(np.float32)

#Print out the 500 person BGL data in mg/dL
for x in range(500):
    print("Person",x+1,"blood glocuse level:",person_BGL_mgdL[x],"mg/dL")
```

# Unit and Data point chooser(more choices inclued)
Section 2 and Section 3 are combined.
Prompt user to input the unit preference and number of data point they want.Print out the requested number of blood glocuse level data in the prefered unit(rounded to 2 decimal place if the unit is mmol) and warn the user if there is an error. 

```python
#Prompt the users to input the preferred unit and the number of data points they want to generate
unit_preference = input("Please type 1 or 2 to choose your preferred unit:\n 1.mmol/L\n 2.mg/dL\n")
number_of_data_point_text = input ("Please enter the number of data points that you want to generate: ")

try:
    #Convert the number_of_data_point_text from string into integer
    number_of_data_point = int(number_of_data_point_text)
    #create a person_BGL array with the number of data requested in floating value lie bewteen 60mg/dL - 128 mg/dL
    person_BGL_mgdL = np.random.randint(60,128,number_of_data_point).astype(np.float32)
    
    #If user prefer mmol/L
    if unit_preference == "1":
        #create a new array person_BGL_mmolL from person_BGL_mgdL
        person_BGL_mmolL = person_BGL_mgdL.copy()
        for x in range(number_of_data_point):
            #Convert the unit from mg/dL to mmol/L and round the value to 2 decimal places
            person_BGL_mmolL[x] = round(person_BGL_mmolL[x]/18,2)
            print("Person",x+1,"blood glocuse level:",person_BGL_mmolL[x],"mmol/L")
    #If user prefer mg/dL
    elif unit_preference == "2":
        for x in range(number_of_data_point):
            print("Person",x+1,"blood glocuse level:",person_BGL_mgdL[x],"mg/dL")
    #If user input other than 1 or 2 in choosing unit
    else:
        print("Please only type 1 or 2 to choose your preferred unit!")
except ValueError: 
    #Output error if user input incorrectly in number of data point
    print("Incorrect input for number of data point")
except:
    print("There is an error")
```
